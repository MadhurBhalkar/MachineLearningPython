```python
def preProcessingText(csv_data,mode):
    tv=TfidfVectorizer(stop_words='english')
    h2=csv_data['Content']
    h2=h2.replace(r'#([^\s]+)', r'\1', regex=True)
    h2=h2.replace('\'"',regex=True)
    content=tv.fit_transform(h2)
    pickle.dump(tv, open("vect.pkl", 'wb'))
    print("Vectorizer File dumped to {}".format(os.path.join(os.getcwd(),'vect.pkl')))


    
    if(mode=='Train'):
        # Modify Labels
        labels = csv_data['Privacy'].astype('category').cat.categories.tolist()
        Label_mapping = {'Privacy' : {k: v for k,v in zip(labels,list(range(1,len(labels)+1)))}}
         
        #Replacing the Strings with Numbers to feed as a input to our model
        
        
        csv_data.replace(Label_mapping, inplace=True)
        return content,csv_data.iloc[:,-1]
    elif(mode=='Predict'):
        return content,None

```

```python
X,Y=preProcessingText(csv1_data,'Train')
```

```python
X_Train, X_Test, Y_Train, Y_Test = model_selection.train_test_split(X, Y, test_size=0.2, random_state=0)
```

```python
def createAndFitModel(X_Train, X_Test, Y_Train, Y_Test):
    model = XGBClassifier()
    model.fit(X_Train, Y_Train)
    predicted = model.predict(X_Test)

    print("Accuracy of the Model is {} %".format(round(model.score(X_Test, Y_Test)*100,2)))
    print("Dumping the file")
    filename = 'XGB_Model.sav'
    pickle.dump(model, open(filename, 'wb'))
    print("File dumped to {}".format(os.path.join(os.getcwd(),'XGB_Model.sav')))
    
createAndFitModel(X_Train, X_Test, Y_Train, Y_Test)
```

```python
# define Predict function 
def predict(X_test):
    #loading vectorizer
    vect_filename=os.path.join(os.getcwd(),'vect.pkl')
    vectorizer = pickle.load(open(vect_filename, "rb"))
    X_test=X_test['Content']
    X_test=X_test.replace(r'#([^\s]+)', r'\1', regex=True)
    X_test=X_test.replace('\'"',regex=True)
    X_test=vectorizer.transform(X_test)
	
    # loading model file
    model_filename = os.path.join(os.getcwd(),'XGB_Model.sav')
    model = pickle.load(open(model_filename, 'rb'))
    Privacy_class_map = {1:'Not Private', 2:'Private'}

	# Test feature
    y_pred = model.predict(X_test)
    y_pred = [round(value) for value in y_pred]
    prediction_result = {'Privacy': Privacy_class_map[y_pred]}
    return prediction_result 

### Lets try it on the new data
#dfNN = pd.DataFrame({'Content': ["b'There is no question that reliable electricity unlocks a better life for people. But can we power the world without\xe2\x80\xa6 https://t.co/TX89IYK8a"]})
#predict(dfNN)
```
